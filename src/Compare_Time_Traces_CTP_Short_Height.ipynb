```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import gainData as gd
import os
import cosmology as cosmo
```

```python
gains=[]
root_dir=('/Users/aaronew/Dropbox_MIT/Science'
          '/simulations-hera-eox/data/timetraces'
          '/varyRiggingHeight_short/')
print root_dir
file_list=[root_dir+'time_trace_run00000%d.txt'%m for m in range(1,6)]
for filename in file_list:
    gains.append(gd.GainData(filename,'CST_TimeTrace',fMin=0.05,fMax=.12))
```

```python
heights=np.array(154.328+np.array([m*(231.492-154.328)/4 for m in range(5)]))/12.*.3048
```

Here we plot the delay-response of the short CTP feed as a function of feed height

```python
for gain,height in zip(gains,heights):
    plt.plot(gain.tAxis,10.*np.log10(np.abs(gain.gainDelay)),label='%.2f meters'%height)
    plt.xlim(-200,500)
    plt.ylim(-60,0)
plt.legend()
plt.axvline(325,color='k',ls='--')
plt.axhline(-55,color='k',ls='--')
plt.grid()
plt.xlabel('$\\tau$ (ns)')
plt.ylabel('gain (dB)')
plt.gcf().set_size_inches(12,6)
plt.title('CTP Delay-Response, 22-inch Cylinder')
```
