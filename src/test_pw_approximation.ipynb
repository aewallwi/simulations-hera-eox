```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from gainData import GainData
```

```python
#load feed only short 
fname_feedonly='../data/timetraces/ctp_connectDisk_feedOnly_timetrace.txt'
gain_feedonly=GainData(fname_feedonly,'CST_TimeTrace',fMin=0.05,fMax=0.13)
fname_feeddish='../data/timetraces/ctp_feedOverDish_timetrace_terminals.txt'
gain_feeddish=GainData(fname_feeddish,'CST_TimeTrace',fMin=0.06,fMax=0.12)
fname_feeddish_pw='../data/timetraces/CTP_feedOverDish_timetrace.txt'
gain_feeddish_pw=GainData(fname_feeddish_pw,'CST_TimeTrace',fMin=0.06,fMax=0.12)
x=gain_feeddish.tAxis
x_pw=gain_feeddish_pw.tAxis-2*4.9*1e9/3e8
y_pw=gain_feeddish_pw.gainDelay
y_ex=gain_feeddish.gainDelay
y_te=gain_feeddish.gain_approx(domain='delay')
y_te_f=gain_feeddish.gain_approx(gain_feedonly.gainFrequency,gain_feedonly.fAxis,domain='delay')
plt.plot(x_pw,10.*np.log10(np.abs(y_pw)/np.abs(y_pw).max()),label='plane wave')
plt.plot(x,10.*np.log10(np.abs(y_te)/np.abs(y_te).max()),label='constant feed approx')
plt.plot(x,10.*np.log10(np.abs(y_te_f)/np.abs(y_te_f).max()),label='full feed sim')
plt.plot(x,10.*np.log10(np.abs(y_ex)),label='excitation')
plt.legend(loc='best')
plt.xlabel('$\\tau$(ns)')
plt.ylabel('$|\widetilde{S}_{11}| (dB)$')
plt.xlim(-100,400)
plt.ylim(-70,0)
```
